```python
#data pulled from json file from covidtracking.com
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
#reading the json file into a dataframe
df = pd.read_json('https://api.covidtracking.com/v1/states/daily.json')
```

```python
#checking the head
df.head()
```

```python
#checking the columns
df.columns
```

```python
#checking the info
df.info()
```

```python
#drop the unimportant columns
df = df.drop(['hash','grade','dateChecked','checkTimeEt','lastUpdateEt','dateModified','negativeTestsAntibody','totalTestsPeopleAntibody',
             'positiveTestsPeopleAntibody','negativeTestsPeopleAntibody','totalTestsPeopleAntigen','positiveTestsPeopleAntigen',
             'totalTestsAntigen','positiveTestsAntigen','fips'],axis=1)   
```

```python
df.columns
```

```python
#can be varied accoridng to the dates we want
df2 = df.loc[df['date'] == 20201001]
```

```python
#creating evenly spaced values
tot_states = np.arange(len(df2['state']))
```

```python
tot_states
```

```python
#California explodes with more number of positive cases
from matplotlib.pyplot import figure
figure(num=None, figsize=(15, 20), dpi=80, facecolor='w', edgecolor='k')
plt.barh(tot_states,df2['positive'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0,0.3,0.3)  )
    
plt.yticks(tot_states, df2['state'])  
plt.xlim(1,max(df2['positive'])) 
plt.xlabel('Positive Number of Cases (Count)')  
plt.title('Corona Virus Positive Cases')  
plt.show()
```

```python
#Again California tops in it as it is a huge state and also had more positive cases
from matplotlib.pyplot import figure
figure(num=None, figsize=(15, 20), dpi=80, facecolor='w', edgecolor='k')
plt.barh(tot_states,df2['negative'], align='center', alpha=0.5,  
                 color=(1,0.2,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(tot_states, df2['state'])  
plt.xlim(1,max(df2['negative'])) 
plt.xlabel('Negative Number of Cases (Count)')  
plt.title('Corona Virus Negative Cases')  
plt.show()
```
